```python
import random
import pandas as pd
import pyterrier as pt 
from pyterrier.measures import *
import os
from pathlib import *
import json
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
cwd = Path(os.getcwd())
path_dataset = os.path.join(cwd.parent.absolute(),'dataset')
path_data_txt = os.path.join(path_dataset, 'data_txt')
path_data_json = os.path.join(path_dataset, 'data_json')

def JsontoDataFrame(profiles_json, split=True):
    rows = []
    docno = 0
    for profile in profiles_json:
        meta_data = {
            "education": json.dumps(profile["Education"]),
            "company": None,
            "title": None,
            "period": None,
        }
        for expBlock in profile["Experience"]:
            meta_data["company"] = expBlock["company"]
            meta_data["title"]   = expBlock["title"]
            meta_data["period"]  = expBlock["period"]
            if expBlock["description"]:
                bulletpoints = expBlock["description"].split('\n') if split else [expBlock["description"]]
                for bulletpoint in bulletpoints:
                    if len(bulletpoint.split()) >= 5:
                        row = [str(docno),bulletpoint]
                        for item in meta_data.values():
                            row.append(item)
                        rows.append(row)
                        docno += 1
    docs_df =   pd.DataFrame(rows,columns= ['docno','text']+list(meta_data.keys()) )
    return docs_df
            
# get raw json files & preprocess
profiles_json = []
for filename in os.listdir(path_data_json):
    path_profile  = os.path.join( path_data_json, filename  )
    with open(path_profile) as ptr:
        profile_json = json.load(ptr)
        profiles_json.append(profile_json)
print(len(profiles_json))
profiles = JsontoDataFrame(profiles_json)
profiles[:10]
        
### buildjson('bullet_points', path_dataset,doc_ls)
```

```python
queries = ['python','software','distributed']
rows = []
for qid,query in enumerate(queries):
    qid += 1
    row = [str(qid), query]
    rows.append(row)
queris = pd.DataFrame(rows, columns=["qid", "query"])
queris[:10]
```

```python
rows = []
for qid,query in enumerate(queries):
    qid += 1
    for docno in list( profiles['docno'] ):
        label = random.randint(0,5)
        iteration = random.randint(1,5)/2
        row = [str(qid),str(docno),int(label),str(iteration)]
        rows.append(row)
qrels = pd.DataFrame(rows, columns=["qid", "docno","label",'iteration'])
qrels[:10]
```

```python
pt_index_path = './resume_index'


### initialization 
if not pt.started():
    pt.init()

### load index fils to index obj
### If index fils are not existed, create index files by indexer
if not os.path.exists(pt_index_path + "/data.properties"):
    index_dir = pt_index_path
    indexer = pt.DFIndexer(index_dir, overwrite=True)
    index_ref = indexer.index(profiles["text"], profiles["docno"])

    ### load the index
    index = pt.IndexFactory.of(index_ref)
else:
    index = pt.IndexFactory.of(pt_index_path)

### show the stat 
print(index.getCollectionStatistics().toString())

### show the reverse index
#for kv in index.getLexicon():
#    print("%s -> %s" % (kv.getKey(), kv.getValue().toString() ) )
print(index.getLexicon()["python"])
```

```python
tfidf = pt.BatchRetrieve(index, wmodel="TF_IDF")
results = tfidf.search("python model nlp")
results[:10]
```

```python
docids = list( results['docid'][:10] )
print(docids)
for docid in docids:
    print( profiles[ profiles['docno'] == str(docid)]['text'] )
```

```python
pt.Experiment(
    [tfidf],
    queris,
    qrels,
    eval_metrics=["map", "ndcg"])
```
